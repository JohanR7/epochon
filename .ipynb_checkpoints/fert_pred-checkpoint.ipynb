```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
import pickle
import os
import datetime

# 1. Load Dataset
df = pd.read_csv('/Users/amanbacker/Desktop/hackathon/data/Fertilizer Prediction 1.csv')

# 2. Separate Features and Labels
features = df[['Temparature', 'Humidity ', 'Moisture', 'Nitrogen', 'Potassium', 'Phosphorous']]
fertilizer_labels = df['Fertilizer Name']

# 3. Encode Labels
fertilizer_encoder = LabelEncoder()
fertilizer_labels = fertilizer_encoder.fit_transform(fertilizer_labels)

# Save encoders for later use
os.makedirs('/Users/amanbacker/Desktop/hackathon/model/pkl_files/', exist_ok=True)
with open("/Users/amanbacker/Desktop/hackathon/model/pkl_files/fertilizer_encoder.pkl", "wb") as file:
    pickle.dump(fertilizer_encoder, file)

# 4. Normalize Features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Save the scaler for later use
os.makedirs('/Users/amanbacker/Desktop/hackathon/model/normalization1/', exist_ok=True)
np.savez("/Users/amanbacker/Desktop/hackathon/model/normalization1/normalization1.npz", mean=scaler.mean_, std=np.sqrt(scaler.var_))

# 5. Split the Data into Training and Validation Sets
X_train, X_val, y_train_fertilizer, y_val_fertilizer = train_test_split(features_scaled, fertilizer_labels, test_size=0.2, random_state=42)

# 6. Define and Compile the Fertilizer Model
fertilizer_model = Sequential([
    Dense(128, activation='relu', input_shape=(features_scaled.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(len(np.unique(fertilizer_labels)), activation='softmax')  # Output layer for Fertilizer
])
fertilizer_model.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 7. Train the Fertilizer Model
EPOCHS = 50
fertilizer_history = fertilizer_model.fit(X_train, y_train_fertilizer, validation_data=(X_val, y_val_fertilizer), epochs=EPOCHS, batch_size=32, verbose=2)

# 8. Save the Trained Fertilizer Model
os.makedirs('/Users/amanbacker/Desktop/hackathon/model/', exist_ok=True)
fertilizer_model_name = str(datetime.datetime.now()).replace(' ', '-').replace(':', '-').replace('.', '-') + '-fertilizer.h5'
fertilizer_model.save(f'/Users/amanbacker/Desktop/hackathon/model/{fertilizer_model_name}')

# 9. Plot the Training and Validation Loss for Fertilizer Model
import matplotlib.pyplot as plt
plt.plot(fertilizer_history.history['loss'], label='Train Loss')
plt.plot(fertilizer_history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Fertilizer Model - Loss')
plt.show()

# 10. Evaluate the Fertilizer Model
fertilizer_train_loss, fertilizer_train_acc = fertilizer_model.evaluate(X_train, y_train_fertilizer, verbose=0)
fertilizer_val_loss, fertilizer_val_acc = fertilizer_model.evaluate(X_val, y_val_fertilizer, verbose=0)

print(f"Fertilizer Model Accuracy on Training Set: {fertilizer_train_acc * 100:.2f}%")
print(f"Fertilizer Model Accuracy on Validation Set: {fertilizer_val_acc * 100:.2f}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import load_model
import pickle

# Load the saved encoder and scaler
with open('/Users/amanbacker/Desktop/hackathon/model/pkl_files/fertilizer_encoder.pkl', 'rb') as file:
    fertilizer_encoder = pickle.load(file)

scaler_data = np.load('/Users/amanbacker/Desktop/hackathon/model/normalization1/normalization1.npz')
scaler_mean = scaler_data['mean']
scaler_std = np.sqrt(scaler_data['std'])
scaler = StandardScaler()
scaler.mean_ = scaler_mean
scaler.scale_ = scaler_std

# Load the trained fertilizer model
fertilizer_model = load_model('/Users/amanbacker/Desktop/hackathon/model/2024-11-30-02-35-33-897790-fertilizer.h5')  # Replace with your actual saved model path

# Function to make a prediction
def predict_fertilizer(nitrogen, phosphorus, potassium, humidity, temperature, missing_feature):
    # Prepare the input features in the same way as the training data
    input_data = np.array([[nitrogen, phosphorus, potassium, humidity, temperature, missing_feature]])  # Added missing feature
    
    # Scale the input data using the same scaler used for training
    input_scaled = scaler.transform(input_data)
    
    # Get the model's prediction
    prediction = fertilizer_model.predict(input_scaled)
    
    # Decode the prediction back to the fertilizer name
    predicted_fertilizer_index = np.argmax(prediction, axis=1)[0]
    predicted_fertilizer = fertilizer_encoder.inverse_transform([predicted_fertilizer_index])[0]
    
    return predicted_fertilizer

# Example array of inputs (now includes 6 features)
input_array = np.array([
    [29.0, 52.0, 45.0, 12.0, 0.0, 36.0],  # Example 1 (added missing feature)
    [40.0, 15.0, 25.0, 70.0, 28.0, 12.0],  # Example 2 (added missing feature)
    [55.0, 25.0, 35.0, 85.0, 32.0, 9.0]    # Example 3 (added missing feature)
])

# Loop through the array and make predictions
for input_data in input_array:
    nitrogen, phosphorus, potassium, humidity, temperature, missing_feature = input_data
    predicted_fertilizer = predict_fertilizer(nitrogen, phosphorus, potassium, humidity, temperature, missing_feature)
    print(f"Input: {input_data} -> Recommended Fertilizer: {predicted_fertilizer}")
```
